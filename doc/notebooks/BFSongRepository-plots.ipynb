```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import article
```

```python
df = article.util.make_df_bf_song_repository()
```

```python
fig, ax = plt.subplots(3, 1)
fig.set_size_inches((10, 20))
ax = ax.ravel()

sns.pointplot(x="test_dir_daynum", y="test_set_dur",
              hue="animal_ID", data=df,
              linestyles=['--' for _ in df.animal_ID.unique()],
              ax=ax[0])
# ax[0].set_ylim([0., 0.1])
ax[0].set_ylabel('test set duration (s)', fontsize=18)
ax[0].set_xlabel('day', fontsize=18)

sns.pointplot(x="test_dir_daynum", y="frame_error_test",
              hue="animal_ID", data=df,
              linestyles=['--' for _ in df.animal_ID.unique()],
              ax=ax[1])
ax[1].set_ylim([0., 0.1])
ax[1].set_ylabel('frame error (%)', fontsize=18)
ax[1].set_xlabel('day', fontsize=18)

sns.pointplot(x="test_dir_daynum", y="syllable_error_test",
              hue="animal_ID", data=df,
              linestyles=['--' for _ in df.animal_ID.unique()],
              ax=ax[2])
ax[2].set_ylim([0., 1.5])
ax[2].set_ylabel('syllable error rate (distance)', fontsize=18)
ax[2].set_xlabel('day', fontsize=18)

fig.suptitle(
    'performance across days\nof models trained with just first 60 seconds of song',
    fontsize=24
);

fig.savefig('../article/figures/fig4/fig4-error-across-days.svg')
fig.savefig('../article/figures/fig4/fig4-error-across-days.png')
```
